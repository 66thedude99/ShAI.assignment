## Task 2
Descriptive Statistics: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
df.describe()
```

```python
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()

print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

## Task 3
Data Cleaning: Handle missing data by suitable method with explain why you use it.

```python
df.isnull().sum()
```

```python
columns_to_replace = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
replacement_value = 0

df[columns_to_replace] = df[columns_to_replace].fillna(replacement_value)
```

```python
columns_to_replace = ['Notes', 'Status']
replacement_value = "NA"

df[columns_to_replace] = df[columns_to_replace].fillna(replacement_value)
```

```python
df.isnull().sum()
```

for null values in the "Salary" columns, i opted to replace with 0
as for null values in the "Notes", "Status" i saw it fits best to replase with "NA"

i also noticed some negative values in my "Salary" columns which i though was weird, but i left them unchanged because i didn't know the context for such values

## Task 4
Basic Data Visualization: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=35, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Base Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
department_counts = df['JobTitle'].value_counts()
department_counts
```

```python
plt.figure(figsize=(10, 8))
top_departments = department_counts.head(20)
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

Some Jobs are the same but with different letter casing, so we're gonna take this into consideration for better results

```python
df['JobTitle'] = df['JobTitle'].str.lower()
department_counts = df['JobTitle'].value_counts()
department_counts
```

```python
plt.figure(figsize=(10, 8))
top_departments = department_counts.head(20) 
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

## Task 5
Grouped Analysis: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
average_salaries_by_job = df.groupby('JobTitle')['TotalPay'].agg(['mean']).reset_index()
average_salaries_by_job
```

```python
average_salaries_by_year = df.groupby('Year')['TotalPay'].agg(['mean']).reset_index()
average_salaries_by_year
```

```python
grouped_data = df.groupby(['JobTitle', 'Year'])['TotalPay'].agg(['mean']).reset_index()

grouped_data
```

```python
top_salaries = average_salaries_by_job.nlargest(50, 'mean')

plt.figure(figsize=(15, 8))
plt.bar(top_salaries['JobTitle'], top_salaries['mean'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Top 50 Average Salaries Across Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.show()
```

```python
lower_salaries = average_salaries_by_job.nsmallest(50, 'mean')

plt.figure(figsize=(15, 8))
plt.bar(lower_salaries['JobTitle'], lower_salaries['mean'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Lower 50 Average Salaries Across Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.show()
```

## Task 6
Simple Correlation Analysis: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

from the get go we can tell there's a correlation between the 'BasePay', 'OvertimePay', 'OtherPay' and the "TotalPay" as it's the result of adding them together.
and the same goes for Benefits', 'TotalPay' as they equal 'TotalPayBenefits' 

```python
import seaborn as sns

numerical_column = 'BasePay'

correlation = df['TotalPay'].corr(df[numerical_column])
print(f'Correlation between TotalPay and {numerical_column}: {correlation:.2f}')

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['TotalPay'], y=df[numerical_column], color='skyblue')
plt.title(f'Scatter Plot: TotalPay vs {numerical_column}')
plt.xlabel('TotalPay')
plt.ylabel(numerical_column)
plt.show()
```

```python
numerical_column = 'TotalPay'

correlation = df['TotalPayBenefits'].corr(df[numerical_column])
print(f'Correlation between TotalPayBenefits and {numerical_column}: {correlation:.2f}')

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['TotalPayBenefits'], y=df[numerical_column], color='skyblue')
plt.title(f'Scatter Plot: TotalPayBenefits vs {numerical_column}')
plt.xlabel('TotalPayBenefits')
plt.ylabel(numerical_column)
plt.show()
```
